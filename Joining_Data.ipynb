```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1 = pd.DataFrame([ {"name" : "furkan", "age" : 26, "location": "ıstanbul"}, 
                   {"name" : "semih", "age" : 33, "location": "ıstanbul"},
                   {"name" : "sibel", "age" : 35, "location" : "malatya"}, 
                      {"name" : "ahmet", "age" : 61, "location" : "malatya"} ])
```

```python
df1
```

```python
df2 = pd.DataFrame( [ { "name" : "sengul", "age" : 55, "location" : "malatya"},
                        {"name": "mehmet", "age" : 45, "location" : "bursa"},
                        {"name" : "veli", "age" : 30, "location": "antalya"},
                        {"name" : "fatma", "age" : 28, "location" : "ıstanbul"}])
```

```python
df2
```

```python
df3 = df1.merge(df2, on = "location")
df3
```

```python
df1.merge(df2, how = "outer")
```

```python
df1.merge(df2, how = "left", on = "location")
```

```python
df1.merge(df2, how = "right", on = "location")
```

```python
df_jan = pd.DataFrame( {"date" : ["2020-01-01", "2020-01-02", "2020-01-03"], "sales" : [100,150,130] } )
df_feb = pd.DataFrame( { "date" : ["2020-02-01", "2020-02-02", "2020-02-03" ], "sales" : [140,160,170] } )
df_march = pd.DataFrame( {"date" : ["2020-03-01", "2020-03-02", "2020-03-03"], "sales" : [90,80,70] } )
```

```python
df_jan
```

```python
df_feb
```

```python
df_march
```

```python
pd.concat([df_jan, df_feb, df_march], ignore_index = True)
```

```python
df10 = pd.concat([df_jan, df_feb, df_march], ignore_index = False, keys = ["jan", "feb", "march"])
df10
```

```python
df_jan.append(df_feb, ignore_index = True)
```

```python
df_feb.append(df_march, ignore_index = True)
```
