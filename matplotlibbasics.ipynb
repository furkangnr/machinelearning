```python
c = np.random.randint(10,50, size = (2,20))
```

```python
df2 = pd.DataFrame(dict(enumerate(c)))
```

```python
plt.figure(figsize = (20,16))
plt.subplot(2,2,1)
plt.scatter(df2[0], df2[1])
plt.subplot(2,2,2)
plt.scatter(df[0], df[1]);
plt.tight_layout()
plt.grid(True)
```

```python
plt.plot(df[0], df[1], marker = "d", color = "green", linestyle = "--", alpha = 0.75)
```

```python
plt.scatter(df[0], df[1], marker = "*", color = "blue", s = 55, alpha = 0.3)
plt.xlim(0,np.max(df[0])+5)
plt.ylim(0,np.max(df[1])+5)
plt.tight_layout()
names_legend = ["first"]
plt.legend(loc ="upper left", labels = names_legend, fontsize = 10)  #upper right , center, lower left , lower right etc...
```

```python
plt.scatter(df[0], df[1], marker = "*", color = "blue", s = 55, alpha = 0.3)
plt.xlim(0,np.max(df[0])+5)
plt.ylim(0,np.max(df[1])+5)
plt.tight_layout()
plt.xlabel("this is x label", fontsize = 15)
plt.ylabel("this is y label", fontsize = 15)
plt.title("this is title", fontsize = 20)
```

## Setting tick marks

```python
plt.scatter(df[0], df[1], marker = "*", color = "blue", s = 55, alpha = 0.3)
plt.xlim(0,np.max(df[0])+5)
plt.ylim(0,np.max(df[1])+5)
plt.tight_layout()
plt.xticks([0,10,20,30,40], ["0", "1A","2A","3A","4A"])
plt.yticks([0,10,20,30,40], ["0", "1B","2B","3B","4B"]);
```

## Working with Image Data

```python
## Matplotlib library only works with .PNG images by default. For further processing with different type of images please check
# Pillow ( PIL ) documentary !!
```

```python
from PIL import Image
```

```python
image_arr = plt.imread("cat.PNG")
```

```python
plt.imshow(image_arr)
```

```python
image_arr.shape
```

```python
# Lets scale the image data :
image_arr2 = image_arr / 2
```

```python
image_arr2[0]
```

```python
image_arr[0]
```

```python
plt.figure(figsize = (15,20))
plt.subplot(2,2,1)
plt.imshow(image_arr2)
plt.subplot(2,2,2)
plt.imshow(image_arr)
```

```python
plt.imsave("cat_scaled.PNG", image_arr2)
```
