```python
image_paths = []
for i in image_files:
    image_paths.append(os.path.join(path, i))
```

```python
images = []
for i in range(len(image_paths)):
    image_ = image.load_img(image_paths[i], target_size = (224,224))
    array_image = image.img_to_array(image_)
    images_rescaled = array_image / 255
    images.append(images_rescaled)
```

```python
default_datagen = ImageDataGenerator( rescale = 1./ 255) 
```

```python
custom_datagen = ImageDataGenerator(featurewise_center = True, rotation_range = 60, width_shift_range = 0.1, 
                                   height_shift_range = 0.1, horizontal_flip = True, vertical_flip = True, rescale = 1./255)
```

```python
transformed_image = default_datagen.apply_transform(x = images_array[1], 
                                                    transform_parameters = {"theta": 45, "tx": 0.25, "ty": 0.15,
                                                                            "brightness" : 0.65, "flip_horizontal" : True,
                                                                            "flip_vertical": True})                                                    
```

```python
plt.imshow(transformed_image / 255)
```

```python
labels = np.zeros(len(images_array))
```

```python
len(labels)
```

```python
iterator = custom_datagen.flow(x = images_array, y = labels, save_to_dir = r"C:\Users\FURKANG\Desktop\Newfolder", 
                               save_prefix = "_")
```

```python
i = 0
created_images = []
created_labels = []
for images, labels in iterator:
    created_images.append(images)
    created_labels.append(labels)
    i = i + 1
    if (i * 32) >  len(images_array):
        break
    
```
